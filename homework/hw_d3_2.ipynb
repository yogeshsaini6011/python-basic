```python
#Check whether two variables point to the same object.
a = input("enter somethingthing")
b= input("enter a somethingthing")
print(f"two variables point to the same object ?? {a is b}" )
```

```python
# Compare two lists using is and ==.
a =[1,2,3,4]
b = [1,2,3,4]
print(f"comparision of {a} with {b} using = is same?? { a==b} ")
print(f"comparision of {a} with {b} using 'is' is same?? { a is b} ")
```

```python
# Demonstrate is vs == using strings.
a = input("enter string ")
b= input("enter string")
print(f"comparision of {a} with {b} using = is same?? { a==b} ")
print(f"comparision of {a} with {b} using 'is' is same?? { a is b} ")

```
