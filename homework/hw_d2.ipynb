## String Slicing (Basics)

```python
#Print the first 4 characters of the string 'Programming
a = 'programing'
print(a[0:4])
```

```python
#Print the last 3 characters of the string 'Python
a = 'python'
print(a[3:6])
```

```python
# From the string 'Developer', extract 'velop' using slicing
a = 'developer'
print(a[2:7])
```

```python
#What will be the output? text = 'Computer' print(text[1:5])
# ompu
```

```python
# Reverse the string 'HelloWorld' using slicing.
a = "hello"
print(a[::-1])
```

## Advanced Slicing

## Mixed Practice (Slicing + Formatting)

```python
#Take a string from the user and print its first half using slicing.
a = input("enter a string")
b  = int(len(a)/2)
print(f"first half of {a} is {a[:b]}")
```

```python
#Take a string from the user, reverse it and display using f-string.
a =input("enter a string")
print(f"reverse of {a} is {a[::-1]}")
```
