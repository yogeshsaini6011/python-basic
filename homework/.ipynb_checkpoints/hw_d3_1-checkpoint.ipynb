##  SET – Programming Questions

```python
#Create a set of 5 numbers and print it.
a= {1,2,3,4,5}
print(a)
```

```python
#Create a set with duplicate values and print the result.
a = {1,2,1,3,4,4,5}
print(a)

```

```python
a=set()
print(a)
```

```python
#Add one element to a set using add().
a = {1.,2,3,4,2}
a.add(6)
print(a)
```

```python
#Add multiple elements to a set using update().
a =  {1.,2,3,4,2}
a.update([1,2,3,4,6])
print(a)

```

```python
# Remove an element from a set using remove()
a = {1.0, 2, 3, 4, 6}
a.remove(6)
print(a)
```

```python
#Remove an element from a set using discard().
a = {1.0, 2, 3, 4, 6}
a.discard(2)
print(a)

```

```python
#Remove a random element from a set using pop().
a = {1.0, 3, 4, 6}
a.pop()
print(a)
```

```python
# Find the length of a set.
a = {1.0, 2, 3, 4, 6}
print(len(a))
```

```python
# Perform union of two sets
a = {1.0, 2, 3, 4}
b = {1.0, 3, 4, 6}
print(a|b)
```

```python
#v  Perform intersection of two sets.
a = {1.0, 2, 3, 4}
b = {1.0, 3, 4, 6}
print(a&b)
```

```python
#2 Find the difference between two sets
a = {1.0, 2, 3, 4}
b = {1.0, 3, 4, 6}
print(a-b)
```

```python
# Find the symmetric difference between two sets.
a = {1.0, 2, 3, 4}
b = {1.0, 3, 4, 6}
print(a^b)
```

##  DICTIONARY – Programming Questions

```python
# Create a dictionary with 3 key-value pairs.
stu = {"name" : "yogesh"
,"roll no" : "11242514"}
print(stu)

```

```python
#Access two values using keys.
stu = {"name" : "yogesh"
,"roll no" : "11242514"}
print(stu ["name"])
print(stu["roll no"])
```

```python
# Add a new key-value pair to a dictionary.
stu['d1'] =12
print(stu)
```

```python
# Update the value of an existing key.
stu['d1']=2
stu
```

```python
#Remove a key using pop(
stu
stu['d']=1
stu['a']=2
stu.pop('name')
stu
```

```python
#Delete a key using del keyword.
del stu['d']
print(stu)
```

```python
#Find the length of a dictionary.
print(len(stu))
```

```python
#Check whether a key exists in a dictionary.
print("name" in stu)
```

```python
print(stu.get("a"))
```

```python
print(stu.keys())
```

```python
#Use values() method.
print(stu.values())
```

```python
#Use items() method.
print(stu.items())
```
