##  List Methods – Programming Questions

```python
#Create a list of 5 fruits and print it.
fruits = ['apple','banana','orange']
print(fruits)
```

```python
# Print the first and last element of a list.

print(f"first element = {fruits[0]} and last elemennt = {fruits[(len(fruits))-1]}")
```

```python
# Print the length of a list.
print(len(fruits))
```

```python
#Create an empty list and add 3 numbers using append().
a = []
a.append('1')
a.append('2')
a.append('3')
print(a)
```

```python
#Add one element to an existing list using append().
fruits.append('mango')
print(fruits)
```

```python
# Insert a value at index 2 in a list using insert()
a.insert(2,'papaya')
print(a)
```

```python
a.insert(0,'python')
print(a)
```

```python
#Create two lists and combine them using extend().
c=[3,4,5,]
c.extend(fruits)
print(c)
```

```python
# Remove a specific value from a list using remove()
a.remove('1')
print(a)
```

```python
# Remove the last element from a list using pop().
c= len(a)
print(f"after removing last element that is {a[c-1]} is {a.pop(c-1)}")
a
```

```python
# Remove an element at index 1 using pop() and print it.
a.pop(1)
print(a)
```

```python
# Sort a list of numbers in ascending order using sort().
p=[1,2,4,5,3,6,8,7,3]
p.sort()

print(p)
```

```python
#Sort a list in descending order using sort().
p=[1,2,4,5,3,6,8,7,3]
p.sort(reverse = True)
print(p)
```

```python
#Use sorted() to sort a list without changing the original list.
c = sorted(p)
c
```

```python
#  Reverse a list using reverse()
p = [1, 2, 3, 3, 4, 5, 6, 7, 8]
p.reverse()
print(p)
```

## Tuple – Programming Questions

```python
# Create a tuple of 5 integers.
p = (1,2,3,4,5)
print(p)
```

```python
#Create a tuple with different data types.
p = (1,2,'a')
print(p)
```

```python
# Create a single-element tuple.
p = (1)
print(p)
```

```python
p = (1, 2, 3, 4, 5)
print(f"first element = {p[0]} and last elemennt = {p[(len(fruits))-1]}")
```

```python
# Access the third element of a tuple.
print(p[3])
```

```python
#Slice a tuple to get the first three elements.
p =(1, 2, 3, 4, 5)
print(p[0:3])
```

```python
# Slice a tuple to get the last two elements.
p =(1, 2, 3, 4, 5)
c=len(p)

print(p[(c-2):c])
```

```python
#Reverse a tuple using slicing.
p =(1, 2, 3, 4, 5)
print(p[::-1])
```

```python
# Count occurrences of an element using count().
p =(1, 2, 3, 4, 5,2,3,4,5)
print(p.count(5))
```

```python
# Find the index of an element using index().
p =(1, 2, 3, 4, 5,2,3,4,5)
print(p.index(5))

```

```python
# Convert a tuple into a list and add one element.
p =(1, 2, 3, 4, 5,2,3,4,5)
c=list(p)
c.append(2342)
p=tuple(c)
print(p)
```
