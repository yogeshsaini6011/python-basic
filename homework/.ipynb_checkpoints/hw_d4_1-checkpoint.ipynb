## Bonus Mini Project

```python
#Student Result System: input name, 5 subject marks, total, percentage, grade, result.
a = int(input("enter marks of sst "))
b = int(input("enter marks of math "))
c= int(input("enter marks of science "))
d = int(input("enter marks of hindi "))
e = int(input("enter marks of english "))
per =((a+b+c+d+e)/500)*100
print(f"total = {a+b+c+d+e}")
print(f"percentage = {((a+b+c+d+e)/500)*100}")
print("result = pass" if per >36 else "result = fail")
if per>36 and per<60 : print("grade = d")
elif per>=60 and per<70 :print("grade = c")
elif per>=70 and per<80 : print("grade = b")
else : print("grade = a")
```
